```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
image = cv2.imread(image_path)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
blurred = cv2.GaussianBlur(image, (5, 5), 0)
blurred_rgb = cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB)
```

```python
edges1 = cv2.Canny(blurred, 50, 100)   # thresholds منخفضة
edges2 = cv2.Canny(blurred, 100, 200)  # thresholds متوسطة
edges3 = cv2.Canny(blurred, 150, 250)  # thresholds أعلى
```

```python
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
plt.imshow(blurred_rgb)
plt.title("Gaussian Blurred Image")
plt.axis("off")
plt.show()
```

```python
plt.imshow(edges1, cmap='gray')
plt.title("Canny Edges (50,100)")
plt.axis("off")
plt.show()
```

```python
plt.imshow(edges2, cmap='gray')
plt.title("Canny Edges (100,200)")
plt.axis("off")
plt.show()
```

```python
plt.imshow(edges3, cmap='gray')
plt.title("Canny Edges (150,250)")
plt.axis("off")
plt.show()
```
